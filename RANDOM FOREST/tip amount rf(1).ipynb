```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('/content/drive/MyDrive/data science/tip-amount_(2).xlsx')
df
```

```python
df.describe()
```

```python
core=df.corr()
plt.subplots(figsize=(8,4))
sns.heatmap(core,vmin=-2,vmax=2,annot=True,linewidth=0.3)
plt.show()
```

```python
df.columns
```

```python
y=df['Observed tip bill(yi)']
x=df.drop(['Observed tip bill(yi)','Meal'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
model=rf.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
prediction
```

```python
print(y_test)
```

```python
print(x_test)
```

```python
print(prediction)
```

```python
from sklearn.metrics import mean_squared_error
print('MSC is:',mean_squared_error(y_test,prediction))
```
