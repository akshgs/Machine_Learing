```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel('/content/drive/MyDrive/data science/tip-amount_(2).xlsx')
df
```

```python
cor=df.corr()
plt.subplots(figsize=(15,5))
sns.heatmap(cor,vmin=-1,vmax=1,annot=True,linewidth=0.2)
plt.show()
```

```python
df.columns
```

```python
y=df['Observed tip bill(yi)']
x=df.drop(['Observed tip bill(yi)','Meal'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
model=dt.fit(x_train,y_train)
```

```python
pre=model.predict(x_test)
pre
```

```python
from sklearn.metrics import mean_squared_error
print('MSE IS :',mean_squared_error(y_test,pre))
```
