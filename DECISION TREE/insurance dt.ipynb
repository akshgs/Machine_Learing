```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('/content/drive/MyDrive/data science/insurance_data.csv')
df
```

```python
df.columns
```

```python
cor=df.corr()
plt.subplots(figsize=(8,4))
sns.heatmap(cor,annot=True,linewidth=0.2)
plt.show()
```

```python
y=df['bought_insurance']
x=df.drop(['bought_insurance'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=45,test_size=0.25)
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
model=dt.fit(x_train,y_train)
```

```python
pre=model.predict(x_test)
pre
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('MSE IS ;',mean_squared_error(y_test,pre))
```
